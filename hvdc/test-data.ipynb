```python
import os
import pandas as pd
import numpy as np
workpath = os.getcwd()
datapath = workpath + "/data/"

# Read csv file
df = pd.read_csv(datapath+"current.csv")
# convert into array data format
X = df.to_numpy()[:,1:].T
print(X.shape)
```

```python
cols = df.columns[1:]
y = np.array([s.split('/')[0] for s in cols])
print(y.shape)
```

```python
import matplotlib.pyplot as plt
plt.plot(X.T)
```

```python
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.gaussian_process.kernels import RBF
kernel = 1.0 * RBF(1.0)
gpc = GaussianProcessClassifier(kernel=kernel,random_state=0).fit(X, y)
gpc.score(X, y)
```

```python
gpc.predict_proba(X)
```
